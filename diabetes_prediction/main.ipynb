```python
def create_features(df):
    
    # Want to create a category for those who smoke and drink
    df["drinker_and_smoker"] = 0
    df.loc[df["Smoker"] == 1 & df["HvyAlcoholConsump"], "drinker_and_smoker"] = 1
    
    # Compiling 
    df["stroke_and_heart_problems"] = 0
    df.loc[df["Stroke"] == 1 & df["HeartDiseaseorAttack"], "previous_complications"] = 1
    
    # Based on the CDC, BMI > 25 is considered overweight. I want to create a column for folks who are larger and don't really move or exercise
    df["immobile_big_body"] = 0
    df.loc[df["DiffWalk"] == 1 & df["BMI"] > 25 & df["PhysActivity"] == 0, "immobile_big_body" ] = 1
    
    
create_features(df)
    
```
